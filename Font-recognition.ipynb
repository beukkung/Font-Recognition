```python
def get_location(path_json,path_img):
    images=[]
    image = cv2.imread(path_img)
    j=open(path_json)
    js=json.load(j)
    for i in range(len(js)):
        try:
            parentId = (js[i]['id'])
            x1 = round(js[i]['x'])
            x2 = round(x1+js[i]['width'])
            y1 = round(js[i]['y'])
            y2 = round(y1+js[i]['height'])
            img =np.array(image[y1:y2,x1:x2])
            w,h=img.shape[0:2]
            if w==h:  img
            elif w>h: img= img[0:h , 0:h]
            elif w<h: img= img[0:w , 0:w]
            img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            img=cv2.resize(img,(32,32))
            img=img[:,:,None]
            images.append(img)
        except:
            img=np.zeros((32,32))
            img=cv2.resize(img,(32,32))
            img=img[:,:,None]
            images.append(img)
    return images

def get_label(path_json):
    parentId=[]
    j=open(path_json)
    js=json.load(j)
    for i in range(len(js)):
        parentId.append(js[i]['id'])
    return parentId
```
